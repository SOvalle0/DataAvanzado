```r
#Función calcular_diferencia
calcular_diferencia <- function(data) {
  data_ordenada <- data[order(data$pib_per_capita), ]

  diferencias <- diff(data_ordenada$pib_per_capita)

  diferencias_df <- data.frame(
    Pais = data_ordenada$Pais[-1],
    Diferencia_PIB = diferencias
  )
  return(diferencias_df)
}
```

Comenta sobre la cantidad de filas en cada dataframe y las posibles implicancias.

```r
#Función filtrar_altos_bajos
filtrar_altos_bajos <- function(data) {
  media_petal_length <- mean(data$Petal.Length, na.rm = TRUE)
  media_sepal_width <- mean(data$Sepal.Width, na.rm = TRUE)

  altos_petal_length <- subset(data, Petal.Length > media_petal_length)

  bajos_sepal_width <- subset(data, Sepal.Width < media_sepal_width)

  return(list(Altos_Petal_Length = altos_petal_length, Bajos_Sepal_Width = bajos_sepal_width))
}
```

```r
#Función sobrevivientes_por_grupo
sobrevivientes_por_grupo <- function(data) {
  resumen <- data %>%
    group_by(Sex, Pclass) %>%
    summarize(Survived_Rate = mean(Survived, na.rm = TRUE) * 100) %>%
    ungroup()
  return(resumen)
}
```

```r
#Aplicación de función sobrevivientes_por_grupo
tasa_supervivencia <- sobrevivientes_por_grupo(titanic)
```

```r
#Visualización de resultados
tasa_supervivencia
```

```r
#Función graficar_supervivencia
graficar_supervivencia <- function(df) {
  ggplot(df, aes(x = factor(Pclass), y = Age, fill = factor(Survived))) +
    geom_boxplot(outlier.shape = NA) +
    geom_jitter(width = 0.2, alpha = 0.4, color = "black") +
    labs(title = "Distribución de la Edad por Clase y Supervivencia",
         x = "Clase de Ticket",
         y = "Edad",
         fill = "Supervivencia") +
    scale_fill_manual(values = c("0" = "red", "1" = "green"),
                      labels = c("No Sobrevivió", "Sobrevivió")) +
    theme_minimal()
}
```

```r
#Aplicación de la función graficar_supervivencia
graficar_supervivencia(titanic)
```
